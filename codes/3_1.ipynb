```python
import pandas as pd
```

```python
data = pd.read_csv("digit recognizor.csv")
```

```python
data.head()
```

```python
X = data.iloc[:, 1:]
Y = data.iloc[:, 0]
```

```python
X.shape
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
selector = VarianceThreshold()
X_var0 = selector.fit_transform(X)
```

```python
X_var0.shape
```

```python
selector.get_feature_names_out()
```

```python
selector.feature_names_in_
```

```python
selected = set(selector.feature_names_in_) - set(selector.get_feature_names_out())
selected
```

```python
sorted(list(selected), key=lambda x : int(x[5:]))
```

```python
len(selected)
```

```python
import numpy as np
```

```python
(X.describe().loc["std", :]**2).median()
```

```python
np.median(X.var().values)
```

```python
x_fsvar = VarianceThreshold(np.median(X.var().values)).fit_transform(X)
```

```python
x_fsvar.shape
```

```python
# 若特征为伯努利分布，则过滤某个分类中占到80%以上的特征
x_bvar = VarianceThreshold(.8 * (1 - .8)).fit_transform(X)
x_bvar.shape
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
cross_val_score(KNN(), X, Y, cv=5).mean()
```

```python
# 方差过滤后
cross_val_score(KNN(), x_fsvar, Y, cv=5).mean()
```

```python
%%timeit
# 随机森林方差过滤前
cross_val_score(RFC(n_estimators=10, random_state=42), X, Y, cv=5).mean()
```

```python
cross_val_score(RFC(n_estimators=10, random_state=42), x_fsvar, Y, cv=5).mean()
```
