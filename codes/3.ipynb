## 归一化

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = [[-1, 2], [-0.5, 6], [0,10], [1,18]]
```

```python
scaler = MinMaxScaler()
scaler.fit(X)

X_scaled = scaler.transform(X)
```

```python
X_scaled
```

```python
# 还愿
X_returned = scaler.inverse_transform(X_scaled)
X_returned
```

```python
# 指定归一化的范围
scaler = MinMaxScaler(feature_range=(5, 10))

X_return_scaled = scaler.fit_transform(X_scaled)
X_return_scaled
```

```python
# 用Numpy
import numpy as np
```

```python
X_array = np.array(X)
X_scaled_numpy = (X_array - X_array.min(axis=0)) / (X_array.max(axis=0) - X_array.min(axis=0))

X_scaled_numpy
```

```python
X_array.max(axis=0)
```

```python
X_scaled_numpy * (X_array.max(axis=0) - X_array.min(axis=0))  + X_array.min(axis=0)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv("Narrativedata.csv", index_col=0)
data.head()
```

```python
data.info()
```

```python
# 实例化类
imp_mean = SimpleImputer()
imp_median = SimpleImputer(strategy="median")
imp_0 = SimpleImputer(strategy="constant", fill_value=0)
```

```python
# 先填补age列
Age = data.loc[:, "Age"].values.reshape(-1,1)
Age.shape
```

```python
imp_mean = imp_mean.fit_transform(Age)
imp_median = imp_median.fit_transform(Age)
imp_0 = imp_0.fit_transform(Age)
```

```python
data.loc[:, "Age"] = imp_median
data.info()
```

```python
embarked = SimpleImputer(strategy="most_frequent")
embarked.fit_transform(data.loc[:, "Embarked"].values.reshape(-1,1))

embarked
```

```python
data.loc[:, "Embarked"] = embarked
data.info()
```

### 编码

```python
from sklearn.preprocessing import LabelEncoder
```

```python
X_copy = data_.copy()
```

```python
# 标签的编码
y = X_copy.iloc[:, -1]

le = LabelEncoder()
le.fit(y)
label = le.transform(y)

le.classes_
```

```python
le.inverse_transform(label)
```

```python
# 取的时sex 和 embarked 列
X_copy.iloc[:, -1] = label
X_copy.head()
```

---

```python
# 分类特征编码为分类数值
from sklearn.preprocessing import OrdinalEncoder
```

```python
X_copy.head()
```

```python
X_copy_ordinal = X_copy.copy()
```

```python
# fit后可以查看每个特征有多少类
OrdinalEncoder().fit(X_copy_ordinal.iloc[:, 1:-1]).categories_
```

```python
# 取出特征列，用ordinal编码
X_copy_ordinal.iloc[:, 1:-1] = OrdinalEncoder().fit_transform(X_copy_ordinal.iloc[:, 1:-1])
```

```python
X_copy_ordinal.head()
```

因此所有的特征都编码为了数字类型

但这种编码方式无法表达变量之间的关系

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
x = X_copy.iloc[:, 1:-1]
```

```python
enc = OneHotEncoder(categories="auto").fit(x)
result = enc.transform(x)
result
```

```python
# 可以发现将特征列OneHot编码后得到一个稀疏矩阵，一共5列，表示总的特征类别之和
# 转为array
result.toarray()
```

```python
pd.DataFrame(enc.inverse_transform(result))
```

```python
# 放入特征矩阵中，并且去掉原来的特征列
X_copy_onehot = pd.concat([X_copy, pd.DataFrame(result.toarray())], axis=1)
X_copy_onehot.head()
```

```python
# 查看每个维度对应的特征取值
enc.get_feature_names_out()
```

```python
X_copy_onehot.drop(["Sex", "Embarked"], axis=1, inplace=True)
X_copy_onehot.head()
```

```python
newX = X_copy_onehot.copy()
newX.columns = ["Age", "Survived", "Female", "Male", "C", "Q", "S"]
newX.head()
```
